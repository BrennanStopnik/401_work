Using the class_heights.txt file and Pandas

1) Clean the data

2) Make column names Title Case

3) Sort gender column by Gender. Female first

4) Convert heights to centimeters

5) Count how many of each gender are in the total class_heights

6) Calc the average height of each gender and the class total

7) Calc the median, mode, and standard deviantion for the total class

```python
import pandas as pd
import numpy as np
```

```python
# 1) Clean the data

frame = pd.read_csv('class_heights.txt',delimiter='|')
frame = pd.DataFrame(frame)
frame
```

```python
# 2) Make column names Title Case

frame.columns = [x.title() for x in frame.columns]
frame
```

```python
# 3) Sort gender column by Gender. Female first

sorted_frame = frame.sort_values(by=['Gender'])
sorted_frame
```

```python
# 4) Convert heights to centimeters

def ht_parse(ht):
    ht_ = ht.split("'")
    ft_ = float(ht_[0])
    in_ = float(ht_[1])
    return round(((12*ft_) + in_) * 2.54,2)

cm_frame = frame['Height'].apply(lambda x: ht_parse(x))

frame['Height'] = cm_frame
frame
```

```python
# 5) Count how many of each gender are in the total class_heights

female = len([x for x in frame['Gender'] if x == 'female'])
male = len([x for x in frame['Gender'] if x == 'male'])

print("\nCount by Gender:")
print(f"Female - {female}")
print(f"Male - {male}")
```

```python
# 6) Calc the average height of each gender and the class total

avg_height_by_gender = frame.groupby('Gender')['Height'].mean()

print("\nAverage height by gender:")
print(avg_height_by_gender)
```

```python
# 7) Calc the median, mode, and standard deviantion for the total class

total_avg_height = frame['Height'].mean()
median_height = frame['Height'].median()
mode_height = frame['Height'].mode()[0]
std_dev_height = frame['Height'].std()

print("\nTotal class average height:")
print(total_avg_height)
print("\nMedian height:")
print(median_height)
print("\nMode height:")
print(mode_height)
print("\nStandard deviation of height:")
print(std_dev_height)
```
