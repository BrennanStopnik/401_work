```python
import pandas
from sqlalchemy import create_engine
```

```python
#1) import CSV -> pandas DF -> Sqlite via Alchemy

# Create the engine to connect to a sqllite database
engine = create_engine("sqlite+pysqlite:///:memory:")

# Read data from CSV, loaded as a dataframe object
data = pandas.read_csv('tips.csv')

# Write data into the table in sqllite database
data.to_sql('tips', engine)
```

```python
data
```

```python
from sqlalchemy import text
with engine.connect() as conn:
    result = conn.execute(text("SELECT * FROM tips"))
    for row in result:
      print(row)
```

```python
#2) SQL database -> table -> Pandas DF via Alchemy
import pandas as pd
from sqlalchemy import create_engine

# SQLAlchemy connectable
cnx = engine.connect()

# table named 'tips' will be returned as a dataframe.
df = pd.read_sql_table('tips', cnx)
df
```
