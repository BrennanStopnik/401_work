Use the previously used yahoo finance pickle files to create two new dataframes.
1) Plot the price of Apple in the month of January 2010 and compare to a graph of the volume of Apple in January 2010.
2) Plot the price of IBM in April 2010 and the price of Miscrosoft in April 2010 on one graph and compare.
3) Plot all prices over the year of 2010 in one graph. This may require some experimentation on your end to see what fits, what doesn't.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt

# Read the pickle files
price = pd.read_pickle('yahoo_price.pkl')
volume = pd.read_pickle('yahoo_volume.pkl')

# Dataframe of price and volume
df_price = pd.DataFrame(price)
df_volume = pd.DataFrame(volume)
```

```python
# 1) Plot the price of Apple in the month of January 2010 and compare to a graph of the volume of Apple in January 2010.


# ax = plt.subplot()

AAPL_price = df_price['AAPL']['2010-01']
plt.title('Price of Apple in January 2010')
# pd.to_datetime(df_price['AAPL']['2010-01'].index)
# plt.xlim(dt.date(2010, 1, 4), dt.date(2010, 1, 29))
plt.xlabel('Date')
plt.ylabel('Price')

AAPL_price.plot(label='Price')
# plt.show()
```

```python
plt.plot(df_volume['AAPL']['2010-01'])
plt.title('Volume of Apple in January 2010')
pd.to_datetime(df_price['AAPL']['2010-01'].index)
plt.xlim(dt.date(2010, 1, 4), dt.date(2010, 1, 29))
plt.xlabel('Date')
plt.ylabel('Volume x 10^8')
plt.show()
```

```python
# 2) Plot the price of IBM in April 2010 and the price of Miscrosoft in April 2010 on one graph and compare.

plt.plot(df_price['IBM']['2010-04'], label='IBM')
plt.title('Price of IBM in April 2010')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
plt.plot(df_price['MSFT']['2010-04'], label='MSFT')
plt.title('Price of MSFT in April 2010')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# 3) Plot all prices over the year of 2010 in one graph. This may require some experimentation on your end to see what fits, what doesn't.

plt.plot(df_price['AAPL']['2010'], label='AAPL')
plt.plot(df_price['IBM']['2010'], label='IBM')
plt.plot(df_price['MSFT']['2010'], label='MSFT')
plt.plot(df_price['GOOG']['2010'], label='GOOG')
plt.title('Price of AAPL, IBM, MSFT and GOOG in 2010')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
