4 - Create pie charts for every Salary Grade with the ratio of men to women

```python
sgs = df_emp_file['sg'].sort_values().unique()

for sg in sgs:
    df_f = df_emp_file['sg'][df_emp_file.sg == sg][df_emp_file['gender'] == 'F'].count()
    print(f'Stats for SG {sg}:')
    print(f'Females: {df_f}')
    df_m = df_emp_file['sg'][df_emp_file.sg == sg][df_emp_file['gender'] == 'M'].count()
    print(f'Males: {df_m} \n')
    sg_pie = pd.DataFrame( {sg: [df_f, df_m]}, index = ['Female', 'Male'])
    sg_pie.plot.pie(y = sg, figsize = (5, 5))

```
